## Imports

```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import tensorflow.keras as keras
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D, BatchNormalization, Flatten, Dropout
```

## Load data

```python
train_dir = 'data/train/'
val_dir = 'data/val/'
test_dir = 'data/test/'

labels_list_str = ['normal', 'pneumonia']
```

```python
def LoadData(data_dir, labels_list_str, img_size = 150):
    images = []
    labels = []
    
    for label_name in labels_list_str:
        label = 0 if label_name == 'normal' else 1
        image_path = os.path.join(data_dir, label_name)
       train_images[0] print(image_path)
        
        for file_name in os.listdir(image_path):
            if file_name.endswith('.jpeg'):
                try:
                    full_path = os.path.join(image_path, file_name)
                    img = cv2.imread(full_path, cv2.IMREAD_GRAYSCALE)
                    img = cv2.resize(img, (img_size, img_size))
                    images.append(img)
                    labels.append(label)
                except Exception as e:
                    print(e)
    return images, np.array(labels)
```

```python
train_images, train_labels = LoadData(train_dir, labels_list_str, img_size = 150)
val_images, val_labels = LoadData(val_dir, labels_list_str, img_size = 150)
test_images, test_labels = LoadData(test_dir, labels_list_str, img_size = 150)
```
